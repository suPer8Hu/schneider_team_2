```python
import boto3
import pandas as pd
from decimal import Decimal
import numpy as np

# Initialize DynamoDB
dynamodb = boto3.resource(
    'dynamodb',
    region_name='us-east-2',

)

def convert_floats_to_decimal(data):
    """Recursively convert float values to Decimal."""
    if isinstance(data, dict):
        return {k: convert_floats_to_decimal(v) for k, v in data.items()}
    elif isinstance(data, list):
        return [convert_floats_to_decimal(v) for v in data]
    elif isinstance(data, float):
        return Decimal(str(data))
    else:
        return data

def upload_csv_to_dynamodb(csv_file, table_name):
    """Upload CSV data to DynamoDB."""
    # Load CSV into DataFrame
    df = pd.read_csv(csv_file)

    # Ensure column names match DynamoDB key names

    # Handle missing or NaN values
    df.fillna('', inplace=True)
    df.replace([np.inf, -np.inf], 1e9, inplace=True)

    # Ensure 'load_id' column exists
    if 'LOAD_ID' not in df.columns:
        raise KeyError("The 'load_id' column is missing from the CSV.")

    # DynamoDB Table
    table = dynamodb.Table(table_name)

    # Upload each row to DynamoDB
    for index, row in df.iterrows():
        item = row.to_dict()
        if not item.get('LOAD_ID'):
            print(f"Skipping row {index + 1}: Missing 'load_id'.")
            continue

        # Convert float values to Decimal
        item = convert_floats_to_decimal(item)

        # Upload to DynamoDB
        try:
            table.put_item(Item=item)
            print(f"Uploaded item {index + 1}: {item}")
        except Exception as e:
            print(f"Failed to upload item {index + 1}: {item}. Error: {e}")

# Usage
csv_file = 'load_posting.csv'
table_name = 'load_postings'
upload_csv_to_dynamodb(csv_file, table_name)
```

```python
import boto3
import pandas as pd
from decimal import Decimal
import numpy as np

# Initialize DynamoDB resource
dynamodb = boto3.resource(
    'dynamodb',
    region_name='us-east-2',

)

def convert_floats_to_decimal(data):
    """Recursively convert float values to Decimal."""
    if isinstance(data, dict):
        return {k: convert_floats_to_decimal(v) for k, v in data.items()}
    elif isinstance(data, list):
        return [convert_floats_to_decimal(v) for v in data]
    elif isinstance(data, float):
        return Decimal(str(data))
    else:
        return data

def upload_csv_to_load_stops(csv_file, table_name):
    """
    Upload data from a CSV file to the `load_stops` DynamoDB table.

    Parameters:
    - csv_file (str): Path to the CSV file.
    - table_name (str): Name of the DynamoDB table.
    """
    # Load the CSV into a DataFrame
    df = pd.read_csv(csv_file)

    # Ensure column names match DynamoDB key names

    # Handle missing or NaN values
    df.fillna('', inplace=True)
    df.replace([np.inf, -np.inf], 1e9, inplace=True)

    # Ensure required columns exist
    if 'LOAD_ID' not in df.columns or 'STOP_ID' not in df.columns:
        raise KeyError("The CSV file must contain 'LOAD_ID' and 'STOP_ID' columns.")

    # Reference to the DynamoDB table
    table = dynamodb.Table(table_name)

    # Iterate through the rows and insert into DynamoDB
    for index, row in df.iterrows():
        # Convert the row to a dictionary
        item = row.to_dict()

        # Ensure required keys are present
        if not item.get('LOAD_ID') or not item.get('STOP_ID'):
            print(f"Skipping row {index + 1}: Missing 'load_id' or 'stop_id'.")
            continue

        # Convert float values to Decimal
        item = convert_floats_to_decimal(item)

        # Upload to DynamoDB
        try:
            table.put_item(Item=item)
            print(f"Uploaded item {index + 1}: {item}")
        except Exception as e:
            print(f"Failed to upload item {index + 1}: {item}. Error: {e}")

# Usage
csv_file = 'load_stop.csv'  # Path to your CSV file
table_name = 'load_stops'  # DynamoDB table name
upload_csv_to_load_stops(csv_file, table_name)
```

```python
import boto3
import pandas as pd

# Initialize DynamoDB resource
dynamodb = boto3.resource(
    'dynamodb',
    region_name='us-east-2',

)

def load_data_from_table(table_name):
    """
    Load data from a DynamoDB table using scan.
    
    Parameters:
        table_name (str): Name of the DynamoDB table.
    
    Returns:
        list: List of items from the table.
    """
    table = dynamodb.Table(table_name)
    data = []
    response = table.scan()

    # Append data from the initial scan
    data.extend(response.get('Items', []))

    # Continue scanning if more data is available
    while 'LastEvaluatedKey' in response:
        response = table.scan(ExclusiveStartKey=response['LastEvaluatedKey'])
        data.extend(response.get('Items', []))
    
    return data

# Load data from DynamoDB tables
load_posting_data = load_data_from_table('load_postings')
load_stop_data = load_data_from_table('load_stops')

# Convert data to pandas DataFrames
posting_data_df = pd.DataFrame(load_posting_data)
stop_data_df = pd.DataFrame(load_stop_data)

# Display information about the loaded data
print(f"Loaded {len(posting_data_df)} records from 'load_posting'")
print(f"Loaded {len(stop_data_df)} records from 'load_stop'")
print(posting_data_df.columns)
print(stop_data_df.columns)
```
